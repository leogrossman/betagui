```python
from epics import PV
freqPV = PV('MCLKHGP:rdFrq')
dlyGenPV = PV('SSMBCCZULP:cell1OnDly')
```

```python
from time import sleep
```

```python
loopactivepv = PV('AKC15VP')
addelaypv = PV('AKC16VP')
```

```python
dg535Delay = 800450 #us
pcDelay = -6.00 #us
currentRF = freqPV.get()
oldDelay = dlyGenPV.get()
print('old Delay:', oldDelay, 'us')

targetDelay = (500000/currentRF * 800000) - dg535Delay + pcDelay
print('new Delay:', targetDelay, 'us')
```

```python
dlyGenPV.put(targetDelay)
```

```python
dg535Delay = 800450 #us
heartbeat = ['/', '-', '\\', '|']
try:
    loopactivepv.put(1)
    i = 0
    while True:
        if loopactivepv.get():
            targetDelay = (500000/freqPV.get() * 800000) - dg535Delay + addelaypv.get()
            print('\r', heartbeat[i%len(heartbeat)], 'new Delay:', targetDelay, 'us', end='')
            dlyGenPV.put(targetDelay)
        sleep(0.5)
        i += 1
finally:
    loopactivepv.put(0)
```
