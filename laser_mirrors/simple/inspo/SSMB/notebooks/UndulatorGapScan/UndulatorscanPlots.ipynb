```python
import numpy as np
from matplotlib import pyplot as plt
import pandas as pd
```

```python
df0 = pd.read_hdf('2024-07-16_U125scan_0.hdf5')
df0
```

```python
scales = [1,1,2.4,1,3,2.7,2.2,3,5,10,1.8,3.2,0.9,1.2,4,4,1.3,1.5]
```

```python
Plot = [3,4,5,6,7]
for i in range(len(df0)//28):
    if i in Plot:
        m = scales[i]
        dfl = df0.loc[i*28:(i+1)*28-1]
        plt.errorbar(dfl.gap, dfl.SSMBavg*m, dfl.SSMBsem*m, capsize=2, marker='', ls='-', label=f'x = {dfl.orbitX.mean():#.3g} mm')
plt.legend()
plt.xlabel('U125 gap / mm')
plt.ylabel('coherent radiation signal / a.u. (arbitrary relative scale)')
plt.show()
```

```python
Plot = [2,3,6,7]
for i in range(len(df0)//28):
    if i in Plot:
        m = scales[i]
        dfl = df0.loc[i*28:(i+1)*28-1]
        plt.errorbar(dfl.gap, dfl.SSMBavg*m, dfl.SSMBsem*m, capsize=2, marker='', ls='-', label=f'x = {dfl.orbitX.mean():#.3g} mm, {dfl.direction.iloc[0]}')
        print(dfl.current.mean())
plt.legend()
plt.xlabel('U125 gap / mm')
plt.ylabel('coherent radiation signal / a.u. (arbitrary relative scale)')
plt.show()
```
