```python
knoblistabs = np.linspace(0,1.4,15)
print(knoblistabs)
```

```python
knoblist = np.concatenate((knoblistabs[:-1], knoblistabs[::-1], -knoblistabs[1:-1], -knoblistabs[::-1]))
print(knoblist)
print(len(knoblist))
```

```python
ksleep = 16
```

```python
step = 0.02
```

```python
for k in knoblist:
    old = knob.get()
    st = step if k > old else -step
    r = np.arange(old,k,st)
    for v in r:
        knob.put(v)
        sleep(0.8)
    knob.put(k)
    sleep(ksleep)
```
