```python
from epics import PV
import numpy as np
```

```python
def pvdiff(pv, diff):
    curset = pv.get()
    pv.put(curset+diff)
```

```python
Q1P1K1RP = PV('Q1P1K1RP:setCur')
Q1P2K1RP = PV('Q1P2K1RP:setCur')
Q1P1L2RP = PV('Q1P1L2RP:setCur')
Q1P2L2RP = PV('Q1P2L2RP:setCur')
Q1P1K3RP = PV('Q1P1K3RP:setCur')
Q1P2K3RP = PV('Q1P2K3RP:setCur')
Q1P1L4RP = PV('Q1P1L4RP:setCur')
Q1P2L4RP = PV('Q1P2L4RP:setCur')

Q1L2 = [Q1P1L2RP,Q1P2L2RP]
Q1rest = [Q1P1K1RP,Q1P2K1RP,Q1P1K3RP,Q1P2K3RP,Q1P1L4RP,Q1P2L4RP]
```

```python
def dispknob2(amount):
    factor = -0.235355 + 0.198 * amount
    scale  = 0.027185
    diffq1l2   = amount * scale
    diffq1rest = amount * scale * factor
    for q1l2 in Q1L2:
        pvdiff(q1l2, diffq1l2)
    for q1rest in Q1rest:
        pvdiff(q1rest, diffq1rest)
        
```
