```python
from time import sleep
```

```python
from epics import PV
```

```python
Q1P1K1RP = PV('Q1P1K1RP:setCur')
Q1P2K1RP = PV('Q1P2K1RP:setCur')
Q1P1L2RP = PV('Q1P1L2RP:setCur')
Q1P2L2RP = PV('Q1P2L2RP:setCur')
Q1P1K3RP = PV('Q1P1K3RP:setCur')
Q1P2K3RP = PV('Q1P2K3RP:setCur')
Q1P1L4RP = PV('Q1P1L4RP:setCur')
Q1P2L4RP = PV('Q1P2L4RP:setCur')
```

```python
Q1ALL  = [Q1P1K1RP, Q1P2K1RP, Q1P1L2RP, Q1P2L2RP, Q1P1K3RP, Q1P2K3RP, Q1P1L4RP, Q1P2L4RP]
Q1RING = [Q1P1K1RP, Q1P2K1RP, Q1P1K3RP, Q1P2K3RP, Q1P1L4RP, Q1P2L4RP]
Q1L2   = [Q1P1L2RP, Q1P2L2RP]
```

```python
def get_list(pvlist):
    return [pv.get() for pv in pvlist]
```

```python
def put_list(pvlist, value):
    for pv in pvlist:
        pv.put(value)
```

```python
def change_list(pvlist, diff):
    for pv in pvlist:
        value = pv.get()
        pv.put(value + diff)
```

```python
get_list(Q1ALL)
```

```python
def old_disp_knob(amount):
    '''
    title= Dispersion bump at ID SSMB250MeV calibrated 1unit=4.85 mm
    '''
    q1l2   = amount *  0.02718475
    q1ring = amount * -0.00630831
    change_list(Q1L2, q1l2)
    change_list(Q1RING, q1ring)
```

```python
def old_disp_angle_knob(amount):
    '''
    title= Dispersion angle bump at ID SSMB250MeV calibrated 1unit= 1.263 mrad
    '''
    q1p1l2val = Q1P1L2RP.get()
    Q1P1L2RP.put(q1p1l2val + amount * 0.0281)
    
    q1p2l2val = Q1P2L2RP.get()
    Q1P2L2RP.put(q1p2l2val - amount * 0.0281)
    
```

```python
for _ in range(30):
    old_disp_knob(amount=-0.03)
    sleep(1)
for _ in range(60):
    old_disp_knob(amount=0.03)
    sleep(3)
for _ in range(30):
    old_disp_knob(amount=-0.03)
    sleep(1)
```

```python
def new_disp_knob(amount):
    q1l2   = amount *  0.02718475
    q1ring = amount * -0.00634220
    change_list(Q1L2, q1l2)
    change_list(Q1RING, q1ring)
```

```python
def new_alpha_knob(amount):
    '''
    amount = Ampere difference on Q1 in ring, Q1_L2 scaled to preserve dispersion at ID
    '''
    q1l2 = amount / 0.2591    
    change_list(Q1L2, q1l2)
    change_list(Q1RING, amount)
```
