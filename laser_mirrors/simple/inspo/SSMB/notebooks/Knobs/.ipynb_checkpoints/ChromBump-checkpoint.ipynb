```python
from epics import PV
from time import sleep
```

bump definition:

S1P1RP:setCur, -4.980

S1P2RP:setCur, -4.980

S2P1RP:setCur, -3.425

S2P2KRP:setCur,-3.425

S2P2LRP:setCur,-3.425 

S3P1RP:setCur,  1.753

S3P2RP:setCur,  1.753

```python
sexts = [PV('S1P1RP:setCur'), PV('S1P2RP:setCur'), PV('S2P1RP:setCur'), PV('S2P2KRP:setCur'), PV('S2P2LRP:setCur'), PV('S3P1RP:setCur'), PV('S3P2RP:setCur')]
sextfactors =  [-4.980, -4.980, -3.425, -3.425, -3.425, 1.753, 1.753]
```

```python
for sext in sexts:
    print(sext.get())
```

```python
def bump_sexts(amount):
    for sext, f in zip(sexts, sextfactors):
        old = sext.get()
        sext.put(old + amount*f)
```

```python
bump_sexts(-0.01)
```
