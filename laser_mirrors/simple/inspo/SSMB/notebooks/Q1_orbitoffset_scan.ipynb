```python
from epics import PV
```

```python
import numpy as np
```

```python
from time import sleep
```

```python
q1offsetpvnames = ['Q1P1K1RP:calcSetCur.G', 'Q1P2K1RP:calcSetCur.G',
               'Q1P1L2RP:calcSetCur.G', 'Q1P2L2RP:calcSetCur.G',
               'Q1P1K3RP:calcSetCur.G', 'Q1P2K3RP:calcSetCur.G',
               'Q1P1L4RP:calcSetCur.G', 'Q1P2L4RP:calcSetCur.G']
```

```python
q1offsetpvs = [PV(n) for n in q1offsetpvnames]
for pv in q1offsetpvs:
    print(pv.get())
```

```python
orbitoffs = PV('AKC12VP')
orbitoffs.get()
```

```python
orange = np.arange(-6,-1.5,.5)
orange
```

```python
qrange = np.arange(0, -0.007, -0.0003)
qrange
```

```python
def set_q1_offset(offset, nsteps=1):
    startingoffsets = [q.get() for q in q1offsetpvs]
    for i in range(nsteps):
        for q, s in zip(q1offsetpvs, startingoffsets):
            q.put(s+ (offset-s)/nsteps*(i+1))
        sleep(0.1)
```

```python
try:
    for o in orange:
        orbitoffs.put(o)
        for q in qrange:
            set_q1_offset(q)
            sleep(3)

finally:
    orbitoffs.put(-5)
    set_q1_offset(0, nsteps=10)
```
