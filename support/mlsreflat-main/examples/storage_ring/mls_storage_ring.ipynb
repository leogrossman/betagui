```python
import at
from mlsreflat.storage_ring import mls_storage_ring, low_alpha, low_emittance, ssmb, injection
```

```python
# Create the ring and print the lattice parameters
ring = mls_storage_ring()
print(at.radiation_parameters(ring))
```

```python
# Set low alpha mode and print the lattice parameters
ring = low_alpha(ring)

print(at.radiation_parameters(ring))
```

```python
# Set low emittance mode and print the lattice parameters
ring = low_emittance(ring)

print(at.radiation_parameters(ring))
```

```python
# Set SSMB mode and print the lattice parameters
ring = ssmb(ring)

print(at.radiation_parameters(ring))
```

```python
# Set injection mode and print the lattice parameters
ring = injection(ring)

print(at.radiation_parameters(ring))
```
